```python
import numpy as np
import matplotlib.pyplot as plt

# data
X = [1.0,2.0,3.0,4.0]
# target
y = [2.0,4.0,6.0,8.0]
```

```python
# weight : taking a random guess
w = 1.0
```

```python
# forward pass for our model
def forward(x):
  # ŷ = x * w
  return x * w
```

```python
# loss function
def loss(x,y):
  y_pred = forward(x)
  # loss = (ŷ-y)²
  return (y_pred-y) * (y_pred-y)
```

```python
weights_list = []
mse_list = []
```

```python
for w in np.arange(0.0,4.1,0.1): # testing w from 0 to 4 taking 0.1 as step
  print(f"w={w}")
  loss_sum = 0 # sum of loss

  for x_val,y_val in zip(X,y): # zip((1,2),(3,4)) => (1,3),(2,4)
     y_pred_val = forward(x_val)
     l = loss(x_val,y_val)
     loss_sum += l
     print(f"x_val: {x_val}, y_val: {y_val}, y_pred_val: {y_pred_val}, l: {l}")

  print(f"MSE = {loss_sum/len(X)}")
  weights_list.append(w)
  mse_list.append(loss_sum/len(X))

```

```python
plt.plot(weights_list,mse_list)
plt.xlabel("Weight")
plt.ylabel("loss")
plt.show()
```
